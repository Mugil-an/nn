```python
import numpy as np
import torch
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def fn(x):
    return 4*x**2- 5*x+2
```

```python
fn(2.0)
```

```python
xs = np.arange(-5,5,0.5)
ys = fn(xs)
print(ys)
```

```python
plt.plot(xs,ys)
```

```python
print(fn(2.0))
h  = 0.00001
print(fn(2.0+h))
```

```python
def derivative(x):
    h = 0.000001
    return (fn(x+h) - fn(x)) / h
```

```python
print(derivative(2.0))
```
